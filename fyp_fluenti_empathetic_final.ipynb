```python
# =========================================
# 3) ROBUST DATASET PROCESSING WITH DEBUGGING - FIXED
# =========================================

print("üì• Loading datasets for COMPREHENSIVE training...")
print(f"üë§ User: hira524")
print(f"üìÖ Current time: 2025-08-22 22:51:25")
print(f"üîß DEBUGGING MODE: Fixing data extraction issues")

# Load datasets
counselchat = load_dataset(
    "json",
    data_files="https://huggingface.co/datasets/Jingy2000/multi-turn-counsel-chat/resolve/main/all_dialogue_cleaned.json"
)["train"]
print(f"   üìä CounselChat total available: {len(counselchat):,} examples")

esconv = load_dataset("thu-coai/ESConv")["train"]
print(f"   üìä ESConv total available: {len(esconv):,} examples")

therapy = load_dataset("vibhorag101/phr_mental_therapy_dataset")["train"]
print(f"   üìä Therapy total available: {len(therapy):,} examples")

mentalchat = load_dataset("ShenLab/MentalChat16K")["train"]
print(f"   üìä MentalChat total available: {len(mentalchat):,} examples")

# Debug: Let's examine the actual data structure
print("\nüîç DEBUGGING: Examining dataset structures...")
if len(counselchat) > 0:
    sample_counsel = counselchat[0]
    print(f"CounselChat sample keys: {list(sample_counsel.keys())}")
    print(f"CounselChat sample: {str(sample_counsel)[:200]}...")

if len(esconv) > 0:
    sample_esconv = esconv[0]
    print(f"ESConv sample keys: {list(sample_esconv.keys())}")
    print(f"ESConv sample: {str(sample_esconv)[:200]}...")

if len(therapy) > 0:
    sample_therapy = therapy[0]
    print(f"Therapy sample keys: {list(sample_therapy.keys())}")
    print(f"Therapy sample: {str(sample_therapy)[:200]}...")

if len(mentalchat) > 0:
    sample_mental = mentalchat[0]
    print(f"MentalChat sample keys: {list(sample_mental.keys())}")
    print(f"MentalChat sample: {str(sample_mental)[:200]}...")

# Define schema for consistency
def create_empty_example():
    return {
        "prompt": "",
        "response": "",
        "source": "",
        "priority": 0.0,
        "valid": False
    }

def create_valid_example(prompt, response, source, priority):
    return {
        "prompt": str(prompt),
        "response": str(response),
        "source": str(source),
        "priority": float(priority),
        "valid": True
    }

# Enhanced context extraction
def extract_comprehensive_therapeutic_context(text):
    if not text or not isinstance(text, str):
        return 'general'

    try:
        text_lower = text.lower()
        contexts = {
            'crisis': ['suicide', 'kill myself', 'end it all', 'no hope', 'can\'t go on'],
            'severe_anxiety': ['panic attack', 'can\'t breathe', 'heart racing', 'going crazy'],
            'trauma': ['flashback', 'triggered', 'abuse', 'assault', 'trauma', 'ptsd'],
            'depression': ['depressed', 'empty', 'numb', 'worthless', 'hopeless'],
            'overwhelm': ['overwhelmed', 'too much', 'drowning', 'chaos', 'can\'t handle'],
            'anxiety': ['anxious', 'worried', 'nervous', 'fear', 'stress'],
            'relationship': ['breakup', 'divorce', 'alone', 'lonely', 'isolated'],
            'academic': ['exam', 'school', 'grades', 'study', 'college']
        }

        for context, keywords in contexts.items():
            if any(keyword in text_lower for keyword in keywords):
                return context
        return 'general'
    except:
        return 'general'

# FIXED: CounselChat Processing with detailed debugging
def process_counselchat_robust(ex):
    try:
        # Debug: Print first few examples to understand structure
        if hasattr(process_counselchat_robust, 'debug_count'):
            process_counselchat_robust.debug_count += 1
        else:
            process_counselchat_robust.debug_count = 1

        if process_counselchat_robust.debug_count <= 3:
            print(f"DEBUG CounselChat {process_counselchat_robust.debug_count}: {ex}")

        # Try multiple possible field names
        question = ""
        answer = ""

        # Common field name variations
        question_fields = ['questionText', 'question', 'input', 'user_input', 'prompt']
        answer_fields = ['answerText', 'answer', 'response', 'output', 'assistant_response']

        for field in question_fields:
            if field in ex and ex[field]:
                question = str(ex[field]).strip()
                break

        for field in answer_fields:
            if field in ex and ex[field]:
                answer = str(ex[field]).strip()
                break

        if not question or not answer:
            return create_empty_example()

        if len(question) < 10 or len(answer) < 15:
            return create_empty_example()

        # Use the original answer instead of generating new one for now
        context = extract_comprehensive_therapeutic_context(question)

        return create_valid_example(
            f"User: {question} Emotion: {context} History: ",
            answer,  # Use original answer
            "counselchat",
            1.0
        )

    except Exception as e:
        if hasattr(process_counselchat_robust, 'error_count'):
            process_counselchat_robust.error_count += 1
        else:
            process_counselchat_robust.error_count = 1
        if process_counselchat_robust.error_count <= 3:
            print(f"DEBUG CounselChat Error {process_counselchat_robust.error_count}: {e}")
        return create_empty_example()

# FIXED: ESConv Processing
def process_esconv_robust(ex):
    try:
        if hasattr(process_esconv_robust, 'debug_count'):
            process_esconv_robust.debug_count += 1
        else:
            process_esconv_robust.debug_count = 1

        if process_esconv_robust.debug_count <= 3:
            print(f"DEBUG ESConv {process_esconv_robust.debug_count}: {ex}")

        # Try to extract from text field
        if 'text' in ex and ex['text']:
            try:
                data = json.loads(str(ex['text']))
                if 'situation' in data:
                    situation = str(data['situation']).strip()
                    if len(situation) >= 10:
                        context = extract_comprehensive_therapeutic_context(situation)
                        # Use a simple empathetic response
                        response = f"I understand this is a difficult situation for you. Your feelings are completely valid. What would be most helpful for you right now?"

                        return create_valid_example(
                            f"User: {situation} Emotion: {context} History: ",
                            response,
                            "esconv",
                            0.8
                        )
            except:
                pass

        return create_empty_example()

    except Exception as e:
        if hasattr(process_esconv_robust, 'error_count'):
            process_esconv_robust.error_count += 1
        else:
            process_esconv_robust.error_count = 1
        if process_esconv_robust.error_count <= 3:
            print(f"DEBUG ESConv Error {process_esconv_robust.error_count}: {e}")
        return create_empty_example()

# FIXED: Therapy Processing
def process_therapy_robust(ex):
    try:
        if hasattr(process_therapy_robust, 'debug_count'):
            process_therapy_robust.debug_count += 1
        else:
            process_therapy_robust.debug_count = 1

        if process_therapy_robust.debug_count <= 3:
            print(f"DEBUG Therapy {process_therapy_robust.debug_count}: {str(ex)[:300]}...")

        # Look for text field
        if 'text' in ex and ex['text']:
            text = str(ex['text'])

            # Simple extraction - look for human/user input
            if 'Human:' in text or 'User:' in text:
                lines = text.split('\n')
                user_line = ""
                assistant_line = ""

                for i, line in enumerate(lines):
                    line = line.strip()
                    if line.startswith('Human:') or line.startswith('User:'):
                        user_line = line.replace('Human:', '').replace('User:', '').strip()
                    elif line.startswith('Assistant:') or line.startswith('AI:'):
                        assistant_line = line.replace('Assistant:', '').replace('AI:', '').strip()

                if user_line and len(user_line) >= 10:
                    if not assistant_line:
                        assistant_line = "I hear what you're sharing with me. This sounds really challenging. What would feel most supportive for you right now?"

                    context = extract_comprehensive_therapeutic_context(user_line)
                    return create_valid_example(
                        f"User: {user_line} Emotion: {context} History: ",
                        assistant_line,
                        "therapy",
                        0.6
                    )

            # Alternative: Look for [INST] format
            elif '[INST]' in text:
                try:
                    start = text.find('[INST]') + 6
                    end = text.find('[/INST]', start)
                    if end > start:
                        user_input = text[start:end].strip()
                        if len(user_input) >= 10:
                            context = extract_comprehensive_therapeutic_context(user_input)
                            response = "I understand you're going through a difficult time. Your feelings are valid and you deserve support. What feels most important to address right now?"

                            return create_valid_example(
                                f"User: {user_input} Emotion: {context} History: ",
                                response,
                                "therapy",
                                0.6
                            )
                except:
                    pass

        return create_empty_example()

    except Exception as e:
        if hasattr(process_therapy_robust, 'error_count'):
            process_therapy_robust.error_count += 1
        else:
            process_therapy_robust.error_count = 1
        if process_therapy_robust.error_count <= 3:
            print(f"DEBUG Therapy Error {process_therapy_robust.error_count}: {e}")
        return create_empty_example()

# FIXED: MentalChat Processing
def process_mentalchat_robust(ex):
    try:
        if hasattr(process_mentalchat_robust, 'debug_count'):
            process_mentalchat_robust.debug_count += 1
        else:
            process_mentalchat_robust.debug_count = 1

        if process_mentalchat_robust.debug_count <= 3:
            print(f"DEBUG MentalChat {process_mentalchat_robust.debug_count}: {ex}")

        # Look for common input field names
        user_input = ""
        assistant_response = ""

        input_fields = ['input', 'question', 'user_input', 'human', 'prompt']
        output_fields = ['output', 'response', 'assistant_response', 'answer']

        for field in input_fields:
            if field in ex and ex[field]:
                user_input = str(ex[field]).strip()
                break

        for field in output_fields:
            if field in ex and ex[field]:
                assistant_response = str(ex[field]).strip()
                break

        if user_input and len(user_input) >= 10:
            if not assistant_response:
                assistant_response = "I can hear that this is difficult for you. Your feelings make complete sense. What kind of support would be most helpful right now?"

            context = extract_comprehensive_therapeutic_context(user_input)
            return create_valid_example(
                f"User: {user_input} Emotion: {context} History: ",
                assistant_response,
                "mentalchat",
                0.4
            )

        return create_empty_example()

    except Exception as e:
        if hasattr(process_mentalchat_robust, 'error_count'):
            process_mentalchat_robust.error_count += 1
        else:
            process_mentalchat_robust.error_count = 1
        if process_mentalchat_robust.error_count <= 3:
            print(f"DEBUG MentalChat Error {process_mentalchat_robust.error_count}: {e}")
        return create_empty_example()

print("\nüîÑ Processing datasets with ROBUST extraction...")

# Process with debugging
print("Processing CounselChat dataset (robust)...")
counsel_processed = counselchat.map(process_counselchat_robust)
counsel_valid = counsel_processed.filter(lambda x: x["valid"])
if len(counsel_valid) > 0:
    counsel_valid = counsel_valid.remove_columns(["valid"])
print(f"‚úÖ CounselChat: {len(counsel_valid):,} valid examples")

print("Processing ESConv dataset (robust)...")
esconv_processed = esconv.map(process_esconv_robust)
esconv_valid = esconv_processed.filter(lambda x: x["valid"])
if len(esconv_valid) > 0:
    esconv_valid = esconv_valid.remove_columns(["valid"])
print(f"‚úÖ ESConv: {len(esconv_valid):,} valid examples")

print("Processing Therapy dataset (robust)...")
# Process only first 1000 for debugging to avoid long processing time
therapy_sample = therapy.select(range(min(1000, len(therapy))))
therapy_processed = therapy_sample.map(process_therapy_robust)
therapy_valid = therapy_processed.filter(lambda x: x["valid"])
if len(therapy_valid) > 0:
    therapy_valid = therapy_valid.remove_columns(["valid"])
print(f"‚úÖ Therapy: {len(therapy_valid):,} valid examples (from sample of 1000)")

print("Processing MentalChat dataset (robust)...")
# Process only first 1000 for debugging
mental_sample = mentalchat.select(range(min(1000, len(mentalchat))))
mental_processed = mental_sample.map(process_mentalchat_robust)
mental_valid = mental_processed.filter(lambda x: x["valid"])
if len(mental_valid) > 0:
    mental_valid = mental_valid.remove_columns(["valid"])
print(f"‚úÖ MentalChat: {len(mental_valid):,} valid examples (from sample of 1000)")

# Enhanced gold standard with more examples
comprehensive_gold_examples = [
    {
        "prompt": "User: I'm struggling with severe anxiety and panic attacks that are affecting my daily life and work performance. Emotion: severe_anxiety History: ",
        "response": "I can hear how difficult and overwhelming this situation is for you. Many people find themselves struggling with anxiety and panic attacks, and your feelings are completely valid. Try taking three slow, deep breaths - in for 4 counts, hold for 4, and out for 6. What kind of support feels most important to you right now?",
        "source": "gold_comprehensive",
        "priority": 1.0
    },
    {
        "prompt": "User: I've been dealing with depression for months and nothing seems to help. I feel empty and worthless most days. Emotion: depression History: ",
        "response": "I understand that you're going through an incredibly challenging time right now. What you're experiencing is a natural response to depression, and these feelings, while painful, are part of your healing journey. Even small acts of self-care, like having a cup of tea or taking a warm shower, can help. What would feel most supportive for you right now?",
        "source": "gold_comprehensive",
        "priority": 1.0
    },
    {
        "prompt": "User: My relationship is falling apart due to constant arguments and I don't know how to fix it. We both love each other but can't communicate properly. Emotion: relationship History: ",
        "response": "I can hear how emotionally complex this relationship situation is for you. These kinds of situations involve many different emotions and communication patterns that can be very challenging to navigate. Given the complexity of relationship dynamics, working with a licensed therapist who specializes in couples therapy would be very beneficial. What aspect of your communication feels most challenging right now?",
        "source": "gold_comprehensive",
        "priority": 1.0
    },
    {
        "prompt": "User: I'm overwhelmed with my final year project, job applications, and family expectations. Everything feels like too much and I can't cope. Emotion: overwhelm History: ",
        "response": "I can sense how much you're carrying right now, and it's completely understandable that you're feeling overwhelmed by all of these important responsibilities. When everything feels too much, try focusing on just getting through the next hour. You don't have to solve everything today - just focus on one small step. What feels like the most urgent priority for you right now?",
        "source": "gold_comprehensive",
        "priority": 1.0
    },
    {
        "prompt": "User: I keep having nightmares about a traumatic experience from my past and I can't sleep properly. Emotion: trauma History: ",
        "response": "I want you to know that what you're experiencing with nightmares is a common response to trauma. Your nervous system is working to process difficult experiences, even though it doesn't feel helpful right now. Given the impact of trauma on sleep and daily functioning, working with a trauma-informed therapist would be very beneficial. What feels most important for your safety and comfort right now?",
        "source": "gold_comprehensive",
        "priority": 1.0
    },
    {
        "prompt": "User: I'm really anxious about my exam results and worried I'll disappoint my family if I don't do well. Emotion: academic History: ",
        "response": "I can hear how much pressure you're feeling about your exam results and your family's expectations. It's completely natural to feel anxious about important outcomes, especially when they matter so much to you and the people you care about. Try to remember that your worth isn't defined by test scores, and your family's love for you goes beyond academic performance. What would help you feel more grounded while waiting for your results?",
        "source": "gold_comprehensive",
        "priority": 1.0
    },
    {
        "prompt": "User: I feel so lonely and isolated since moving to a new city. I don't know how to make friends and connect with people. Emotion: relationship History: ",
        "response": "I understand how difficult and isolating it can feel to start over in a new place. Moving somewhere new is a big life change, and it's completely normal to feel lonely while you're building new connections. Many people struggle with making friends as adults - you're definitely not alone in this experience. What are some activities or interests you enjoy that might help you meet like-minded people?",
        "source": "gold_comprehensive",
        "priority": 1.0
    }
]

gold_dataset = Dataset.from_list(comprehensive_gold_examples)

# Combine all available datasets
datasets_to_combine = [gold_dataset]  # Always include gold standard

if len(counsel_valid) > 0:
    datasets_to_combine.append(counsel_valid)
    print(f"üìä Including CounselChat: {len(counsel_valid):,} examples")

if len(esconv_valid) > 0:
    datasets_to_combine.append(esconv_valid)
    print(f"üìä Including ESConv: {len(esconv_valid):,} examples")

if len(therapy_valid) > 0:
    datasets_to_combine.append(therapy_valid)
    print(f"üìä Including Therapy: {len(therapy_valid):,} examples")

if len(mental_valid) > 0:
    datasets_to_combine.append(mental_valid)
    print(f"üìä Including MentalChat: {len(mental_valid):,} examples")

# Combine datasets
combined_comprehensive = concatenate_datasets(datasets_to_combine)

print(f"\n‚úÖ ROBUST DATASET PROCESSING COMPLETE")
print("=" * 60)
print(f"üìä Total examples ready for training: {len(combined_comprehensive):,}")
print(f"üìÖ Dataset prepared: 2025-08-22 22:51:25")
print(f"üéØ Ready for comprehensive therapeutic training!")

if len(combined_comprehensive) > 0:
    print(f"\nüìù Sample robust response:")
    sample = combined_comprehensive[0]
    print(f"Input: {sample['prompt'][:100]}...")
    print(f"Response: {sample['response'][:200]}...")
    print(f"Source: {sample['source']}")
```

```python
# =========================================
# 4) COMPREHENSIVE THERAPEUTIC TOKENIZATION
# =========================================

print(f"üî§ Comprehensive tokenization setup - 2025-08-22 22:51:25")
print(f"üë§ Training for user: hira524")
print(f"üìä Processing {len(combined_comprehensive):,} comprehensive examples")

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

# Enhanced professional therapeutic system message
COMPREHENSIVE_SYSTEM_MESSAGE = """You are a professional, empathetic mental health support assistant trained on comprehensive therapeutic data. Your responses follow evidence-based therapeutic standards:

CORE THERAPEUTIC PRINCIPLES:
‚Ä¢ Validation & Empathy: Acknowledge their experience with genuine understanding
‚Ä¢ Normalization: Help them feel less alone in their struggles
‚Ä¢ Strength Recognition: Identify their resilience and coping efforts
‚Ä¢ Grounding Techniques: Offer practical, immediate coping strategies
‚Ä¢ Hope & Direction: Provide realistic optimism and next steps
‚Ä¢ Collaborative Language: Use "we" and "together" approaches
‚Ä¢ Professional Boundaries: Support without diagnosis, refer when appropriate
‚Ä¢ Cultural Sensitivity: Respect diverse backgrounds and experiences

RESPONSE STRUCTURE:
1. Validate their experience with empathy
2. Normalize their feelings/situation
3. Offer practical support or coping strategy
4. Ask one thoughtful, open-ended question
5. End with hope and strength recognition

SPECIALIZED RESPONSES:
- Crisis situations: Immediate safety focus + professional referral
- Complex relationships: Family/couples therapy referral + support
- Trauma indicators: Trauma-informed language + specialized care referral
- Mental health symptoms: Validation + coping techniques + professional guidance

Trained on comprehensive data for hira524's professional mental health chatbot - 2025-08-22 22:51:25"""

def tokenize_comprehensive_therapeutic(batch):
    """Comprehensive tokenization with enhanced system message"""
    conversations = []
    for prompt, response in zip(batch["prompt"], batch["response"]):
        # Comprehensive therapeutic conversation format
        conversation = f"""{COMPREHENSIVE_SYSTEM_MESSAGE}

{prompt}

Comprehensive Therapeutic Assistant: {response}"""
        conversations.append(conversation)

    return tokenizer(
        conversations,
        truncation=True,
        max_length=MAX_LEN,
        padding="max_length",
        return_tensors=None
    )

print("‚ö° Tokenizing comprehensive therapeutic dataset...")
print(f"üìä Processing {len(combined_comprehensive):,} comprehensive examples...")

tokenized_comprehensive = combined_comprehensive.map(
    tokenize_comprehensive_therapeutic,
    batched=True,
    remove_columns=["prompt", "response", "source", "priority"],
    desc="Tokenizing comprehensive therapeutic data"
)

tokenized_comprehensive = tokenized_comprehensive.shuffle(seed=42)

# Comprehensive train/eval split
split = tokenized_comprehensive.train_test_split(test_size=0.1, seed=42)  # Larger eval set
train_ds, eval_ds = split["train"], split["test"]

print(f"\nüè• COMPREHENSIVE TRAINING DATA READY")
print("=" * 55)
print(f"üìö Training samples: {len(train_ds):,}")
print(f"üìä Evaluation samples: {len(eval_ds):,}")
print(f"üìè Max sequence length: {MAX_LEN} tokens")
print(f"‚è±Ô∏è Estimated training time: 60-90 minutes (comprehensive)")
print(f"üéØ Target: Comprehensive therapeutic response quality")
print(f"üìà Scale: Large-scale professional training")
print(f"üë§ Custom trained for: hira524")
print("=" * 55)
```

```python
# =========================================
# 10) ADVANCED COMPREHENSIVE RESPONSE SYSTEM
# =========================================

if training_success:
    print(f"üß† Initializing Advanced Comprehensive Response System...")
    print(f"üë§ User: hira524")
    print(f"üìÖ Current time: 2025-08-22 23:05:53")

    def clean_and_enhance_comprehensive_response(raw_response, user_input=""):
        """Advanced post-processing for comprehensive therapeutic responses"""

        # Remove system message artifacts
        response = raw_response
        if COMPREHENSIVE_SYSTEM_MESSAGE.split('\n')[0] in response:
            response = response.split(COMPREHENSIVE_SYSTEM_MESSAGE.split('\n')[0])[-1]

        if "Comprehensive Therapeutic Assistant:" in response:
            response = response.split("Comprehensive Therapeutic Assistant:")[-1]

        # Clean formatting
        response = response.strip()
        response = re.sub(r'\n+', ' ', response)  # Replace newlines
        response = re.sub(r'\s+', ' ', response)  # Multiple spaces to single

        # Professional quality control
        sentences = response.split('.')

        # Ensure professional structure: validation + technique + question + hope
        if len(sentences) > 8:  # Too long, keep essential parts
            # Find key components
            validation_sent = next((s for s in sentences if any(word in s.lower() for word in ['understand', 'hear', 'feel', 'know'])), sentences[0])
            technique_sent = next((s for s in sentences if any(word in s.lower() for word in ['breathe', 'ground', 'focus', 'moment'])), "")
            question_sent = next((s for s in sentences if '?' in s), "")
            hope_sent = next((s for s in sentences if any(word in s.lower() for word in ['heal', 'strength', 'hope', 'change'])), sentences[-1])

            key_parts = [s for s in [validation_sent, technique_sent, question_sent, hope_sent] if s.strip()]
            response = '. '.join(key_parts) + '.'

        # Question control (max 2 questions)
        question_count = response.count('?')
        if question_count > 2:
            parts = response.split('?')
            response = '?'.join(parts[:2]) + '?'

        # Ensure professional ending
        if not response.endswith(('.', '!', '?')):
            response += '.'

        return response

    class ComprehensiveTherapeuticBot:
        def __init__(self, model, tokenizer, config):
            self.model = model
            self.tokenizer = tokenizer
            self.config = config
            self.conversation_history = []
            self.session_start = datetime.utcnow()
            self.user = "hira524"

        def detect_crisis_indicators(self, user_input):
            """Detect crisis-level indicators requiring immediate attention"""
            crisis_keywords = [
                'suicide', 'kill myself', 'end it all', 'want to die',
                'no point living', 'better off dead', 'can\'t go on'
            ]
            return any(keyword in user_input.lower() for keyword in crisis_keywords)

        def generate_comprehensive_response(self, user_input, emotion="general", max_length=180):
            """Generate comprehensive therapeutic response with quality control"""

            # Crisis detection
            if self.detect_crisis_indicators(user_input):
                emotion = "crisis"
                print("‚ö†Ô∏è Crisis indicators detected - using specialized response")

            # Enhanced prompt with comprehensive context
            comprehensive_prompt = f"""{COMPREHENSIVE_SYSTEM_MESSAGE}

User: {user_input} Emotion: {emotion} History: {self._get_recent_history()}

Comprehensive Therapeutic Assistant:"""

            # Tokenize with comprehensive standards
            inputs = self.tokenizer(
                comprehensive_prompt,
                return_tensors="pt",
                truncation=True,
                max_length=MAX_LEN-80  # Leave room for comprehensive response
            ).to(self.model.device)

            # Generate with comprehensive parameters
            with torch.no_grad():
                outputs = self.model.generate(
                    inputs.input_ids,
                    attention_mask=inputs.attention_mask,
                    max_new_tokens=max_length,
                    do_sample=True,
                    temperature=0.6,           # More controlled for comprehensive consistency
                    top_p=0.9,                # Focused but natural
                    top_k=50,                 # Professional vocabulary focus
                    repetition_penalty=1.2,    # Reduce repetition
                    pad_token_id=self.tokenizer.eos_token_id,
                    eos_token_id=self.tokenizer.eos_token_id,
                    early_stopping=True
                )

            # Decode and clean response
            full_response = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

            # Extract assistant response
            if "Comprehensive Therapeutic Assistant:" in full_response:
                response = full_response.split("Comprehensive Therapeutic Assistant:")[-1].strip()
            else:
                response = full_response[len(comprehensive_prompt):].strip()

            # Comprehensive post-processing
            final_response = clean_and_enhance_comprehensive_response(response, user_input)

            # Quality validation
            if len(final_response.split()) < 30:
                # Fallback to ensure minimum comprehensive quality
                final_response = create_professional_therapeutic_response_fixed(user_input, emotion)

            # Log interaction
            self.conversation_history.append({
                "timestamp": datetime.utcnow().strftime('%Y-%m-%d %H:%M:%S UTC'),
                "user_input": user_input,
                "detected_emotion": emotion,
                "response": final_response,
                "crisis_detected": self.detect_crisis_indicators(user_input),
                "response_length": len(final_response.split())
            })

            return final_response

        def _get_recent_history(self):
            """Get recent conversation context for continuity"""
            if not self.conversation_history:
                return ""

            recent = self.conversation_history[-2:]  # Last 2 interactions
            history_text = ""
            for interaction in recent:
                history_text += f"Previous: {interaction['user_input'][:50]}... "

            return history_text.strip()

        def get_session_summary(self):
            """Generate session summary for comprehensive documentation"""
            return {
                "user": self.user,
                "session_start": self.session_start.strftime('%Y-%m-%d %H:%M:%S UTC'),
                "session_end": datetime.utcnow().strftime('%Y-%m-%d %H:%M:%S UTC'),
                "total_interactions": len(self.conversation_history),
                "crisis_interactions": sum(1 for i in self.conversation_history if i['crisis_detected']),
                "avg_response_length": sum(i['response_length'] for i in self.conversation_history) // max(1, len(self.conversation_history)),
                "emotions_addressed": list(set(i['detected_emotion'] for i in self.conversation_history))
            }

        def clear_session(self):
            """Clear session history"""
            self.conversation_history = []
            self.session_start = datetime.utcnow()

    print("‚úÖ Advanced Comprehensive Response System Ready!")
    print("üè• Features: Crisis detection, quality control, session tracking")
    print("üéØ Comprehensive therapeutic standards maintained")

else:
    print("‚ùå Skipping advanced response system due to training failure")
```
